```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv')
df
```

```python

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(df['Mileage'], df['Age(yrs)'], df['Sell Price($)'], color = "green")
plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```

```python
dummy = pd.get_dummies(df['Car Model'])
merged = pd.concat([df,dummy],axis='columns')
final = merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
X = final.drop('Sell Price($)',axis='columns')
y = final['Sell Price($)']
```

```python
reg.fit(X,y)
```

```python
reg.predict([[45000,4,0,0]])
```

```python
reg.predict([[86000,7,0,1]])
```

```python
reg.score(X,y)
```
